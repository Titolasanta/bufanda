```python
import pandas as pd
import math as mt
import seaborn as sns
import unidecode
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("frag/events_of_search engine hit.csv",index_col = 0)
```

```python
df.head()
```

```python
eventos = df["search_engine"].value_counts()

sns.set(rc={'figure.figsize':(14.6,8.2)})
g = sns.barplot(x = eventos.values, y = eventos.index)
g.set_title("Top 5 de empresas", fontsize=15)
g.set_xlabel("Cantidad de Hits", fontsize=14)
g.set_ylabel("Empresas de advertising", fontsize=14)
```

```python
df["timestamp"] = pd.to_datetime(df["timestamp"])
```

```python
df["day"] = df["timestamp"].dt.day
df["mes"] = df["timestamp"].dt.month
```

```python
agraf = df["day"].value_counts().sort_index()
agraf.plot(kind = 'line', figsize = (16, 8), title = 'Search engine hits por dia', legend = True, fontsize = 12)
```

```python
aplot = df.groupby("person").size().value_counts()
plot = aplot.plot(kind = 'line', figsize = (16, 8), title = 'cantidad de personas con cada cantidad de hits', legend = True, fontsize = 12)

plot.set_ylabel("cantidad de personas")
plot.set_xlabel( "cantidad de hits de las personas")
```
